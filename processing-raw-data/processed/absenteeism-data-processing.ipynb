Dữ liệu cột có dấu hiệu bất thường khi cột có giá trị min là 0, max là 28, song dữ liệu dù tăng dần từ 0 tới 28 lại chỉ có độ dài là 28 (lẽ ra là 29 do từ 0 tới 28 có 29 số).

Ngoài ra, việc phân loại data dựa trên lý do (từ 0 tới 28) hiện tại đang không cho thấy ý nghĩa mà chỉ đơn thuần là định danh cho từng lý do. Do đó, chúng ta sẽ phân loại lý do nghỉ theo danh mục và sử dụng dummy variables (tạo ra một var nếu chứa giá trị 1 là True, 0 là False).

<b> 1. Tạo Dummy Variables

```python
reason_columns = pd.get_dummies(df['Reason for Absence'])
```

```python
reason_columns
```

Do lý do 0 (giá trị thuộc cột 'Reason for Absence' có giá trị = 0) không xuất hiện trong bảng, ta sẽ drop cột này.

```python
reason_columns = reason_columns.drop([0], axis = 1)
```

Cột ở dạng True/False có thể gây khó khăn cho việc tính toán, định lượng về sau. Do đó, ta chuyển cột về dạng int (số 0 cho False, 1 cho True)

```python
reason_columns = reason_columns.astype(int)
```

```python
reason_columns
```

<b> 2. Drop cột "Reason for Absence"

Do đã chuyển những lý do vắng sang dạng dummy variables, bây giờ ta có thể drop cột "Reason for Absence" để dataset gọn hơn.

```python
df = df.drop(['Reason for Absence'], axis = 1)
```

Dựa theo Data Dictionary, có thể phân nhóm các lý do từ 1 tới 28 theo 4 nhóm chính, cụ thể: 
- <b> Nhóm A </b> là các lý do liên quan tới bệnh lý và rối loạn sức khỏe, bao gồm các lý do từ 1 tới 14. 
- <b> Nhóm B </b> là các lý do liên quan tới sức khỏe sinh sản và bẩm sinh, bao gồm các lý do từ 15 tới 17. 
- <b> Nhóm C </b> là các lý do liên quan tới triệu chứng và nguyên nhân ngoại lai, bao gồm các lý do từ 18 tới 21. 
- <b> Nhóm D </b> là các lý do liên quan tới yếu tố cá nhân và môi trường, bao gồm các lý do còn lại.

Hiện tại các bảng lý do vẫn đang ở dạng nhiều cột, trong khi để đưa vào bảng lớn, chúng ta sẽ chỉ cần 1 cột để xác định lý do vắng mặt. Điều này là do mỗi nhân viên sẽ chỉ vắng mặt vì một lý do (xin vắng vì một lý do). Tuy nhiên, đây chỉ là giả thiết ngộ nhận, chúng ta cần kiếm tra để chắc chắn rằng mỗi nhân viên chỉ có một lý do vắng mặt thông qua việc tạo thêm cột check và cộng tổng theo dòng.

Lý giải cho việc (cộng tổng) này, vì dữ liệu ban đầu là thống kê vắng mặt của nhân viên theo ngày, do đó mỗi ngày một nhân viên sẽ chỉ xin vắng vì một lý do. Nếu có hơn 2 lý do xin nghỉ trong một ngày, có thể dữ liệu đang có vấn đề.

```python
reason_columns['check'] = reason_columns.sum(axis=1)
```

```python
reason_columns['check'].unique()
```

Cột chỉ trả về 2 giá trị là 0 và 1 (có thể nghỉ việc hoặc không). Do đó không có vấn đề về dữ liệu, ta có thể bỏ cột 'check' khỏi bảng.

```python
reason_columns = reason_columns.drop(['check'], axis=1)
```

```python
reason_columns
```

Sau khi đã xác nhận mỗi nhân viên chỉ nghỉ việc vì một lý do, ta tiến hành đua các cột về theo nhóm, các nhóm về một cột duy nhất.

```python
reasons_A = reason_columns.loc[:, 1:14].max(axis=1)
reasons_B = reason_columns.loc[:, 15:17].max(axis=1) 
reasons_C = reason_columns.loc[:, 18:21].max(axis=1) 
reasons_D = reason_columns.loc[:, 22:].max(axis=1)
```

<b> 1. Chuyển kiểu dữ liệu

Các giá trị cột 'Date' đang ở giá trị string - chuỗi, do đó cột đang ở sai định dạng. Cần đổi cột về định dạng chính xác.

```python
checkpoint_df['Date'] = pd.to_datetime(checkpoint_df['Date'], format = '%d/%m/%Y')
```

```python
type(checkpoint_df['Date'][0])
```

<b> 2. Chuyển về dạng ý nghĩa

Cột 'Date' hiện ở dạng có cả ngày, tháng, năm. Nếu phân tích vắng mặt, cột sẽ trở nên ý nghĩa hơn nếu ta đưa về các phân cấp thời gian nhỏ hơn. Ví dụ nếu ngữ cảnh dữ liệu xảy ra tại Việt Nam thì các tháng cuối năm, giáp Tết như tháng 1, tháng 12 có thể chứng kiến lượng vắng mặt tăng đột biến. Hoặc lượng vắng mặt cũng có thể tăng đột biến nếu ngày vắng mặt rơi vào một thứ Hai, đến từ tâm lý chán nản sau 2 ngày nghỉ cuối tuần, hoặc thứ Sáu, với tâm lý sắp nghỉ các ngày cuối tuần.

Từ phân tích sơ bộ này, cột 'Date' có thể trở nên ý nghĩa hơn nếu ta chuyển về dạng Tháng hoặc Thứ trong tuần.

```python
month_lists = []
```

```python
for i in range(len(checkpoint_df['Date'])): 
    month_lists.append(checkpoint_df['Date'].loc[i].month)
```

```python
month_lists
```

Thêm cột tháng vào dataframe.

```python
checkpoint_df['Month Value'] = month_lists
```

```python
checkpoint_df
```

Ta tiến hành quy trình tương tự với cột thứ trong tuần.

```python
def day_in_week(date):
    return date.weekday()
```

```python
checkpoint_df['Day In Week Value'] = checkpoint_df['Date'].apply(day_in_week)
```
